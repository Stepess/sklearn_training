# Basic machine leanring algorithm using Sklearn lib

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

# Data
## Wine dataset

```python
class_name = 'Cultivars'

feature_cols = [
        'Alcohol',
        'Malic_acid',
        'Ash',
        'Alcalinity_of_ash',
        'Magnesium',
        'Total_phenols',
        'Flavanoids',
        'Nonflavanoid_phenols',
        'Proanthocyanins',
        'Color_intensity',
        'Hue',
        'OD280/OD315_of_diluted_wines',
        'Proline'
    ]

cols = feature_cols.copy()
cols.insert(0, class_name)

df = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data', 
                 header=None, names=cols)

df.head()
```

```python
df.info()
```

# Missing data

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

# Visualization

```python
plt.figure(figsize=(15,15))
ax = sns.heatmap(df.corr(), 
            annot=True, 
            vmin=-1,
            cmap='coolwarm')

ax.set_xticklabels(ax.get_xticklabels(), rotation=45, fontsize=12, rotation_mode='anchor', ha='right')
ax.set_yticklabels(ax.get_yticklabels(), fontsize=12)
```

```python
fig, axs = plt.subplots(figsize=(15,5), ncols=3)
sns.distplot(df['Alcohol'], kde=False, color='brown', bins=15, ax=axs[0])
sns.distplot(df['Ash'], kde=False, color='yellow', bins=15, ax=axs[1])
sns.distplot(df['Proline'], kde=False, color='blue', bins=15, ax=axs[2])
```

```python
fig, axs = plt.subplots(figsize=(15, 7), ncols=2)
sns.boxplot(x='Cultivars', y='Alcohol', data=df,palette='winter', ax=axs[0])
sns.boxplot(x='Cultivars', y='Ash', data=df, ax=axs[1])
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

features = df.drop(class_name, axis=1)
target = df[class_name]

scaled_features = scaler.fit_transform(features)

df_feat = pd.DataFrame(scaled_features, 
             index=features.index,
             columns=features.columns)

df_feat.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

x_train, x_test, y_train, y_test = train_test_split(df_feat, target, test_size=0.3)

def report(actual, predicted):
    print('Accuracy: {0}\n'.format(accuracy_score(actual, predicted)))
    print('Confusion matrix:\n\n {0}\n'.format(confusion_matrix(actual, predicted)))
    print('Classification report:\n\n {0}'.format(classification_report(actual, predicted)))
```

# k Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier

error_rate = []

for i in range(1,30):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    pred_i = knn.predict(x_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,30),error_rate,color='blue', marker='o',
         markerfacecolor='red', markersize=10)
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
knn = KNeighborsClassifier(n_neighbors=19)
```

```python
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
report(y_test, y_pred)
```

# Decision tree

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()
dtc.fit(x_train, y_train)
```

```python
y_pred = dtc.predict(x_test)
report(y_test, y_pred)
```

```python
from IPython.display import Image  
from sklearn.externals.six import StringIO  
from sklearn.tree import export_graphviz
import pydot 
```

```python
dtc_visualization = StringIO()  
export_graphviz(dtc, out_file=dtc_visualization,feature_names=feature_cols,filled=True,rounded=True)

graph = pydot.graph_from_dot_data(dtc_visualization.getvalue())  
Image(graph[0].create_png())  
```

# Support Vector Machines

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(x_train, y_train)
```

```python
y_pred = svc.predict(x_test)
print('Accuracy: {0}'.format(accuracy_score(y_test,y_pred)))
```

```python
from sklearn.model_selection import GridSearchCV

svc_param = {
        'kernel': ['linear', 'rbf'],
        'gamma': [0.0001, 0.001, 0.01, 0.1, 0.5, 1.0, 10.0],
        'C': [0.1, 1, 10, 100, 1000]
    }

grid = GridSearchCV(estimator=svc,
                  param_grid=svc_param,
                  n_jobs=-1,
                  refit=True,
                  verbose=3,
                  cv=5)
```

```python
%%time
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
svc = grid.best_estimator_
```

```python
y_pred = svc.predict(x_test)
report(y_test, y_pred)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(x_train, y_train)
```

```python
y_pred = rfc.predict(x_test)
print('Accuracy: {0}'.format(accuracy_score(y_test,y_pred)))
```

```python
rfc_param = {
    'max_depth': [80, 100, None],
    'max_features': [2, 3, 'auto'],
    'n_estimators': [100, 200, 300, 1000]
}

grid = GridSearchCV(estimator=rfc,
                  param_grid=rfc_param,
                  n_jobs=-1,
                  refit=True,
                  verbose=3,
                  cv=5)
```

```python
%%time
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
rfc = grid.best_estimator_
```

```python
y_pred = rfc.predict(x_test)
report(y_test, y_pred)
```

```python
feature_imp = pd.Series(rfc.feature_importances_, index=feature_cols).sort_values(ascending=False)
print(feature_imp)
sns.barplot(x=feature_imp, y=feature_imp.index)
plt.xlabel('Feature Importance Score')
plt.ylabel('Features')
plt.title("Visualizing Important Features")
plt.legend()
plt.show()
```

# Find the best

```python
from sklearn.metrics import f1_score

algorithms = ['kNN', 'DecesionTree', 'SVM', 'RandomForest', 'AdaBoost', 'GradientBoosting']

f1_scores = [
        f1_score(y_test, knn.predict(x_test), average='macro'),
        f1_score(y_test, dtc.predict(x_test), average='macro'),
        f1_score(y_test, svc.predict(x_test), average='macro'),
        f1_score(y_test, rfc.predict(x_test), average='macro'),
        f1_score(y_test, abc.predict(x_test), average='macro'),
        f1_score(y_test, gbc.predict(x_test), average='macro'),
    ]

alg_f1 = pd.DataFrame({
    'algorithms': algorithms,
    'f1': f1_scores
})

alg_f1 = alg_f1.sort_values('f1', ascending=False)

alg_f1 = alg_f1.reset_index(drop=True)

alg_f1
```
